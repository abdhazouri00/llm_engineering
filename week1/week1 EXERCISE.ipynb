```python
from IPython.display import display, Markdown, update_display

def chatbot():
    # 1. Initialize the conversation history with the system prompt
    history = [
        {"role": "system", "content": system_prompt}
    ]

    print("ğŸ“ Tutor is ready! (Type 'exit' to stop)")

    while True:
        # 2. Get user input
        user_input = input("\nYou: ")
        
        # 3. Check for exit command
        if user_input.lower() in ["exit", "quit"]:
            print("Goodbye!")
            break

        # 4. Add user message to history
        history.append({"role": "user", "content": user_input})

        # 5. Create the stream with the FULL history
        stream = openai.chat.completions.create(
            model="gpt-4o-mini", # Changed to a real model name (gpt-5-nano isn't public yet)
            messages=history,
            stream=True
        )

        # 6. Set up the display for this specific turn
        response_content = ""
        display_handle = display(Markdown(data="â³ Thinking..."), display_id=True)

        for chunk in stream:
            token = chunk.choices[0].delta.content or ""
            response_content += token
            # Update the same display handle to create the "typing" effect
            update_display(Markdown(response_content), display_id=display_handle.display_id)

        # 7. Add the full assistant response to history so it remembers it next time
        history.append({"role": "assistant", "content": response_content})
```
